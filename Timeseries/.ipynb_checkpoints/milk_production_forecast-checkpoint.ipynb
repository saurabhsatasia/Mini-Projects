```python
import pandas as pd
import fbprophet
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv("data/monthly-milk-production-pounds.csv")
```

```python
df.sample(5)
```

```python
# drop row/index number 168 with NaN
df.drop(168, axis=0, inplace=True)
```

```python
df.plot()
```

### 1. FbProphet requires date column named as "ds"
### 2. FbProphet requires output column named as "y"

```python
df.columns=["ds","y"]
df.sample(2)
```

```python
# convert "ds" to datetime format
df["ds"] = pd.to_datetime(df["ds"])
df.sample(2)
```

```python
# # #### Step 1. Convert seasonal data to stationary data
# # * shift 1 position down to convert to stationary, first row data will be Nan.
# # * If data does not look stationary ny 1 shift, shift by 2 and so on, unless data doeas not look stationary
# # * We can also use DickiFuller Test, check the value of p, unless p value is not less than 0.05, data is not considering stationary
# df["y"] = df['y']-df['y'].shift(1) # shift 1 position down, first position will Nan
# df.head()
```

```python
df['y'].plot()
```

```python
from fbprophet import Prophet
print(dir(Prophet))
```

```python
# initialize the Model
model = Prophet()
model.fit(df)
```

For forecasting , we have to create some future dates

```python
# Create future dates for 365 days
future_dates = model.make_future_dataframe(periods=365)
```

```python
df.tail(2)
```

```python
future_dates
```

```python
# prediction
prediction = model.predict(future_dates)
prediction.head()
```

```python
prediction[['ds','yhat','yhat_lower','yhat_upper']].tail()
```

#### Plotting the Forecasts
Prophet has an inbuilt feature that enables us to plot the forecasts we just generated. This is achieved using model.plot() and passing in our forecasts as the argument. The blue line in the graph represents the predicted values while the black dots represents the data in our dataset.

```python
# plot
model.plot(prediction)
```

```python
# Visualize each componets [Trends, Weekly]
model.plot_components(prediction)
```

#### Cross Validation
Next lets measure the forecast error using the historical data. Well do this by comparing the predicted values with the actual values. In order to perform this operation we select cut of points in the history of the data and fit the model with data upto that cut off point. Afterwards we compare the actual values to the predicted values. The cross_validation method allows us to do this in Prophet. This method take the following parameters as explained below:

1. horizon the forecast horizon
2. initial the size of the initial training period
3. period the spacing between cutoff dates

```python
from fbprophet.diagnostics import cross_validation

df_cv=cross_validation(model, horizon="365 days", period="180 days", initial="1095 days")
```

```python
df_cv.head()
```

```python
# 1964, 1968
# 1964, 1965 -->Train data
# 1965, 1966 --> Test data
# 1966, 1968 --> Validation
```

#### Obtaining the Performance Metrics
We use the performance_metrics utility to compute the Mean Squared Error(MSE), Root Mean Squared Error(RMSE),Mean Absolute Error(MAE), Mean Absolute Percentage Error(MAPE) and the coverage of the the yhat_lower and yhat_upper estimates.

```python
from fbprophet.diagnostics import performance_metrics
df_performance = performance_metrics(df_cv)
df_performance.head()
```

```python
from fbprophet.plot import plot_cross_validation_metric
fig = plot_cross_validation_metric(df_cv, metric='rmse')
```

```python
fig = plot_cross_validation_metric(df_cv, metric='mse')
```

```python
fig = plot_cross_validation_metric(df_cv, metric='mae')
```
