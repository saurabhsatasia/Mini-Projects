```python
import pandas as pd
import fbprophet
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv("data/monthly-milk-production-pounds.csv")
```

```python
df.sample(5)
```

```python
# drop row/index number 168 with NaN
df.drop(168, axis=0, inplace=True)
```

```python
df.plot()
```

### 1. FbProphet requires date column named as "ds"
### 2. FbProphet requires output column named as "y"

```python
df.columns=["ds","y"]
df.sample(2)
```

```python
# convert "ds" to datetime format
df["ds"] = pd.to_datetime(df["ds"])
df.sample(2)
```

#### Step 1. Convert seasonal data to stationary data
* shift 1 position down to convert to stationary, first row data will be Nan.
* If data does not look stationary ny 1 shift, shift by 2 and so on, unless data doeas not look stationary
* We can also use DickiFuller Test, check the value of p, unless p value is not less than 0.05, data is not considering stationary

```python
df["y"] = df['y']-df['y'].shift(1) # shift 1 position down, first position will Nan
df.head()
```

```python
df['y'].plot()
```

```python
from fbprophet import Prophet
print(dir(Prophet))
```

```python
# initialize the Model
model = Prophet()
model.fit(df)
```

For forecasting , we have to create some future dates

```python
# Create future dates for 365 days
future_dates = model.make_future_dataframe(periods=365)
```

```python
df.tail(2)
```

```python
future_dates
```

```python
# prediction
prediction = model.predict(future_dates)
prediction.head()
```

```python
prediction[['ds','yhat','yhat_lower','yhat_upper']].tail()
```

```python
# plot
model.plot(prediction)
```

```python
# Visualize each componets [Trends, Weekly]
model.plot_components(prediction)
```

#### Cross Validation

```python
from fbprophet.diagnostics import cross_validation

df_cv=cross_validation(model, horizon="365 days", period="180 days", initial="1095 days")
```

```python
df_cv.head()
```

```python
from fbprophet.diagnostics import performance_metrics
df_performance = performance_metrics(df_cv)
df_performance.head()
```
